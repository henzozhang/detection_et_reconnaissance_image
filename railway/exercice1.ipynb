```python
import matplotlib.pyplot as plt
from skimage.io import imread,imshow
from skimage.color import rgb2gray  
from skimage import exposure
import numpy as np
```

```python
image1 = imread('railway.jpg')
```

```python
image_grey = rgb2gray(image1)
imshow(image_grey)
```

```python
image_grey =imread('railway.jpg', as_gray=True)
imshow(image_grey)
```

```python
image_grey[1,1]
```

```python
im = np.where(image_grey > 150/255, 0, 1)

imshow(im, cmap=plt.get_cmap('gray'))
```

```python
def histGrayScale(img):
    _,axes = plt.subplots(2,2, figsize=(12,6))
    axe=axes.ravel()
    axe[0].imshow(img, cmap=plt.get_cmap('gray'))
    axe[0].set_title('Image')  
    hist =exposure.histogram(img)
    axe[1].plot(hist[1])
    axe[2].plot(hist[0])
    
histGrayScale(image_grey)
```

```python
histGrayScale(im)
```

```python
def histcolorScale(img):
    _,axes = plt.subplots(ncols=2, figsize=(12,6))
    
    axes[0].imshow(img)
    axes[0].set_title('Image')  
    axes[1].set_title('Histogram')
    axes[1].plot(exposure.histogram(img[...,0])[0],color='red')
    axes[1].plot(exposure.histogram(img[...,1])[0],color='green')
    axes[1].plot(exposure.histogram(img[...,2])[0],color='blue')

histcolorScale(image1)
```

```python
img_rose = imread('R.jpg')
histcolorScale(img_rose)
```

```python
img_b = imread('B.jpg')
histcolorScale(img_b)
```

```python
img_tulip = imread('tulip.jpg')
histcolorScale(img_tulip)
```
