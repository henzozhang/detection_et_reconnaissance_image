```python
photo_rouge=np.zeros((photo2.shape[0]//2,photo2.shape[0]//2,3))
photo_vert=np.zeros((photo2.shape[0]//2,photo2.shape[0]//2,3))
photo_blue=np.zeros((photo2.shape[0]//2,photo2.shape[0]//2,3))
photo_original=np.zeros((photo2.shape[0]//2,photo2.shape[0]//2,3))
for i in range(photo2.shape[0]//2):
    for j in range(photo2.shape[1]//2):
        photo2[i][j][0]=0
        photo_rouge[i][j]=photo2[i][j]
for i in range(photo2.shape[0]//2,photo2.shape[0]):
    for j in range(photo2.shape[1]//2):
        photo2[i][j][1]=0

for i in range(photo2.shape[0]//2):
    for j in range(photo2.shape[1]//2,photo2.shape[1]):
        photo2[i][j][2]=0

        
photo_rouge[:,:,:]=photo2[:photo2.shape[0]//2,:photo2.shape[1]//2,:]
photo_vert[:,:,:]=photo2[photo2.shape[0]//2:,:photo2.shape[1]//2,:]
photo_blue[:,:,:]=photo2[:photo2.shape[0]//2,photo2.shape[1]//2:,:]
photo_original[:,:,:]=photo2[photo2.shape[0]//2:,photo2.shape[1]//2:,:]
plt.subplot(2,2,1)
plt.imshow(photo_rouge)
plt.subplot(2,2,2)
plt.imshow(photo_vert)
plt.subplot(2,2,3)
plt.imshow(photo_blue)
plt.subplot(2,2,4)
plt.imshow(photo_original)
```
